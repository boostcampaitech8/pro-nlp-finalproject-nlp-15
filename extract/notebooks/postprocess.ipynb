```python
import os
import json
import re

# 1. 수정 함수 정의
def fix_simple_tail(text):
    if not isinstance(text, str):
        return text
    # 문장 끝에 '다'와 '.'이 섞여 반복되는 구간을 '다.' 하나로 치환
    return re.sub(r'(다[\s\.]*)+$', '다.', text.strip())

# 2. 경로 설정
version = 'v4_3d'
events_path = '../../outputs/events'
input_folder = f'{events_path}/{version}/result'   # 원본 jsonl 파일들이 있는 폴더
output_folder = f'{events_path}/{version}/postprocessed'   # 수정된 파일이 저장될 폴더

# 출력 폴더가 없으면 생성
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
    print(f"폴더 생성 완료: {output_folder}")

# 3. 파일 처리 루프
files = [f for f in os.listdir(input_folder) if f.endswith('.jsonl')]

for filename in files:
    input_path = os.path.join(input_folder, filename)
    output_path = os.path.join(output_folder, filename)
    
    with open(input_path, 'r', encoding='utf-8') as infile, \
         open(output_path, 'w', encoding='utf-8') as outfile:
        
        print(f"처리 중: {filename}...")
        
        for line in infile:
            if not line.strip():
                continue
                
            # JSON 로드
            data = json.loads(line)
            
            # description 필드 수정 (존재할 경우)
            if 'description' in data:
                data['description'] = fix_simple_tail(data['description'])
            
            # 한 줄씩 새 파일에 쓰기
            json_line = json.dumps(data, ensure_ascii=False)
            outfile.write(json_line + '\n')

print(f"\n모든 작업이 완료되었습니다! 결과물 확인: {output_folder}")
```
